<h1>
Balanced Binary Search Tree ?
</h1>

BST의 경우, 높이가 h인 경우 값을 탐색하고 삭제하고 삽입하는데 O(h)만큼의 시간이 필요했다. 
최악의 경우 연결리스트와 동일한 모양으로 나열될 경우, n에 비례하여 시간이 걸린다. 
BST 연산은 전체 트리의 높이가 매우 큰 영향을 미친다. 
n개의 노드가 있는 경우 트리가 가질 수 있는 최대의 높이는 n-1이지만, 
자식을 둘씩 꽉 채운 완전 이진트리의 형태로 구성하는 경우 최대 높이는 logn이 된다. (밑이 2인 로그) 

아하! 트리를 꽉꽉 채워서 만들면 언제나 높이는 logn보다 작게되고, 그렇다면 연산 속도도 logn에 비례하겠구나!

그런데 기존 트리에서의 높이는 삽입하는 순서에 따라서 영향을 받았다. (언제나 꽉 꽉 채워서 이진트리가 형성되지 않았다 젠장)
예를 들어 BST에서 1->2->3->4 순서로 삽입하는 경우 높이는 3이지만, 3->2->4->1 순서의 경우 높이는 2이다. 

아하 그렇다면 삽입 순서와 관계없이, 언제나 최소한의 높이를 갖도록 BST를 구성할 수 있을까? 
=> 이것이 바로 Balanced BST다.

Balanced BST란, 탐색시간이(높이가) O(logN)인 BST를 모두 일컫는 말이다.
최소한의 높이를 유지하도록 삽입과 삭제 시 트리를 재구성하는 자료구조다. 
이때, 회전이라는 방식을 통해 트리를 재구성한다.




<h1>
AVL Tree, Adelson-Vesky, Landis Tree (1964)
</h1>

B-BST는 여러 종류가 있다. 그중 AVL은 최초의 균형 이진탐색트리다. 높이가 logn에 비례하도록 하는 규칙은 종류마다 다르다.
AVL의 규칙은 다음과 같다. 

<h3>
" 모든 노드에 대해서 
노드의 왼쪽 sub-tree와 오른쪽 sub-tree의 
높이차가 1이하로 유지되도록 설계된 BST이다. "
</h3>

이를 유지하기 위해 BalanceFactor를 사용한다. 이는 양쪽 자식 노드의 높이 차를 의미하며, right - left이다.
노드 v를 기준으로 왼쪽 subTree의 높이가 오른쪽보다 하나 더 높으면, v의 BalanceFactor는 -1이다. 
AVL 트리의 정의에 따라, 모든 노드의 BalanceFactor는 -1, 0, 1 중에 하나를 갖는다. 

HOW 어떻게 높이차를 1이하로 유지하는가? => 언제 높이 차이가 발생하나?
삽입과 삭제 시 높이차가 1보다 커질 수 있다. 
따라서 이런 경우마다 균형이 깨지는 지점에 대해서 rotate를 한 번 또는 두 번 수행하여 높이를 재조정한다.

<h1>
AVL Class
</h1>
- Rules
    1) 모든 노드에 대한 left와 right 의 높이차가 1이하가 되도록 한다.
    2) 삽입, 삭제 시 시 균형이 깨질 수 있는 지점을 찾고, root까지 올라가며 재구성
    3) 재구성은 rotation이라는 연산을 통해 진행
        - left와 right rotation이 있고 서로 대칭적이다. 두 연산을 한번씩 진행하면 원래대로 돌아온다.
        - 회전 후에도 BST 값을 순회한 결과가 그대로 유지되어야한다.

- Attributes

- Methods
    Special Methods
    `rotationRight`     
    `rotationLeft`
    `rebalance(x, y, z) ` 
        1) 균형 깨진 녀석 z를 기준으로 삽입(삽입된 노드까지) 자식 y, 손자 x에 대해서 생길 수 있는 모양은 4가지이다.
            => x, y, z에 대해 일직선 배치인 경우 => 한 번의 rotate 필요
                예시 : (left, left, z) 또는 (right, right, z)
            => x, y, z에 대해 삼각형 배치인 경우 => 두 번의 rotate 필요
                예시 : (right, left, z) 또는 (left, right, z)
        2) rebalance() 이후 기존 z위치에 있던 노드를 반환해야한다. => 기존 z가 root라서 그 정보를 갱신해야할찌도 모르니까 
        3) 균형이 깨진 최초의 노드에 대해서만 수행하면 된다. 왜냐면 원래 AVL이었는데 insert나 delete로 균형이 깨진 거니까. 한 칸의 높이만 수정하면 끝ㅋ
    `isBalanced`
    `insertAVL` 
        1) BST클래스의 insert 호출한다. 삽입된 노드 v를 반환한다. 
        2) v부터 root 까지 올라가면서 조상들을 조사하여 처음으로 균형이 깨진 노드 z를 기준으로 x, y, z를 찾는다. 
            => BalanceFactor의 크기가 2 또는 -2가 된다. (노드 하나가 생겼으니까 3은 될 수 없다.)
                => 삽입된 노드 v가 균형이 깨진 z의 손자 x일 수도 있다. 하지만, 삽입된 v가 y일 순 없다. 왜냐면 그럼 BalanceFactor가 깨지지 않기 때문
                => z기준으로 오른자식은 없고, 왼자식으로 v가 생겼다면 BalanceFactor는 -1, 반대의 경우 1
                => 기타 등등..  
        3) rebalance(x, y, z) 호출한다. 반환한 값을 w라고 할 때,
        4) w가 root인지 조사하고, 그렇다면 root 정보를 갱신한다.

    `deleteAVL`
        1) BST의 delete 중 하나를 호출하고, 균형이 깨질 수 있는 노드 v를 반환한다. 
        2) x, y, z를 찾는다. insert와 다르게, 올라가며 찾는 게 아니라, 균형이 깨진 노드로부터 양쪽 서브트리와의 노드를 비교하며 더 깊은 쪽으로 y와 x를 선택한다.
        3) 하지만 삽입에선 균형을 한번만 맞추면 끝났었다. 
        하지만 삭제의 경우, 아래에서 균형을 맞췄지만 부모 입장에선 다시 균형이 깨질 수 있다. 따라서 root까지 올라가면서 계속 균형을 맞춰야한다. 
</pre>

```python
from my_modules.tree.my_BST_v2 import BST

class AVL(BST):
    
    # 오른쪽으로 한칸씩 재배치 => 부모 z, 오른자식 c, 왼자식 x, x왼자식 a, x오른자식 b   
    def rotateRight(self, z):
        # step1. 부모z자리에 자식x를 대신 갖다놓는다. (x와 z가 모두 존재한다면)
        # 재구성할 기준이 되는 노드 자체가 존재하는지 확인
        if z == None: return
        x = z.left    
        # 기준의 왼쪽 자식 노드 확인
        if x == None: return
        # b는 나중에 z의 오른쪽 자식인, c의 왼쪽 자식으로 붙일 예정
        b = x.right
        #z의 부모를 x가 차지함
        x.parent = z.parent
        # 근데 z가 root가 아닌 경우에 대해서
        if z.parent:
            #그럼 z가 원래 왼쪽자식이었는지, 오른쪽 자식이었는지 알아야 x를 갖다 붙이니까
            if z.parent.left == z:
                z.parent.left = x
            else:
                z.parent.right = x
        # 만약 z가 root였다면? root를 다시 지정해줄 필요가 있음
        else: 
            self.root = x
        # 현재까지 z자리에 x를 앉힘, z는 기존의 부모와의 연결이 끊이진 상테(새로운 부모와의 연결이 필요)
        x.right = z
        z.parent = x
        # 그리고 x의 오른쪽 요소로 기존의 b를 내버리고, z가 앉았으니, b는 x의 왼팔로 들어감
        z.left = b
        if b: 
            b.parent = z
        #x와 z의 높이를 업데이트 =>b는 높이 변화없이 z의 왼자식으로 들어갔고, 그런 녀석을 새롭게 자식으로 뒀으니 z를 검사하고, 
        # z의 높이가 변했다면, x에 영향을 미치니, z가 균형이 꺠지는 지점임
        self.updateHeight(z)
        
    # rotateRight를 원상 복구하는 느낌이랄까 현재 x의 왼자식 a, 오른자식 z, z왼자식 b, z오른자식 c
    def rotateLeft(self, x):
        # 현재 x를 왼쪽으로 밀어내고, 그 자리를 z가 차지해야하는 상황이므로 x와 z가 무조건 존재해야함
        if x == None: return
        z = x.right
        if z == None: return
        b = z.left
        # x 의 자리를 뺴앗음
        z.parent = x.parent
        # x가 root가 아닌 경우, 
        if x.parent:
            # 새로운 z를 왼자식으로 둘지 오른자식으로 둘지 결정
            if x.parent.left == x:
                x.parent.left = z
            else:
                x.parent.right = z
        # x가 root였다면, 새로운 root 할당
        else: 
            self.root = z
        #자리를 빼앗은 z는 x를 자신의 왼쪽 자식으로 삼음
        z.left = x
        x.parent = z
        # 이로 인해서 기존의 z의 왼팔이었던, b는 x의 오른팔로 들어감( 존재한다면)
        x.right = b
        if b:
            b.parent = x
        self.updateHeight(x)

    # 균형이 깨진 지점(x)까지의 노드 x(손자) - y(자식) - z(부모)
    # 모양에 따라서 rotate를 1회 또는 2회 호출하여 한 칸의 높이 차를 줄임
    def rebalance(self, x, y, z):
        # 균형이 깨진 노드가 없다면 재배치할 필요 없음
        if z == None: pass
        
        # 일직선인 경우
        # (left, left, z)
        if z.left == y and y.left == x:
            # 왼쪽으로 기울어진 상태 => 오른쪽으로 회전
            self.rotateRight(z)
            # z의 자식이던 y가 z자리를 차지함, root가 될 경우를 대비하여 반환
            # print("### AVL.rebalance : 일직선인 경우 (left, left, z)", y)
            return y
        # (right, right, z)
        elif z.right == y and y.right == x:
            # 오른쪽으로 기울어진 상태 => 왼쪽으로 회전
            self.rotateLeft(z)
            # print("### AVL.rebalance : 일직선인 경우 (right, right, z)", y)
            return y
        # 삼각형인 경우
        # (right, left, z)
        elif z.right == y and y.left == x:
            # 오른으로 뾰족한 상태, 꺾인 부분을 펼쳐서 일직선만든 뒤 rotate
            # y기준으로 왼쪽으로 꺾임
            self.rotateRight(y) 
            # 현재  y, x, z가 (right, right, z)로 배치됨
            self.rotateLeft(z)
            # 가장 상위에 x가 와있으니까, 
            # print("### AVL.rebalance : 삼각형인 경우 (right, left, z)", x)
            return x
        # (left, right, z)
        elif z.left == y and y.right == x:
            # 왼쪽으로 뾰족한 상태, 오른쪽으로 돌아간 y를 펴줘
            self.rotateLeft(y)
            # (left, left, z)상태라서 오른쪽으로 돌림
            self.rotateRight(z)
            # print("### AVL.rebalance : 삼각형인 경우 (left, right, z)", x)
            return x

    # 1) 일단 BST하게 삽입 후 해당 결과를 v에 저장
            # v가 None인 경우 => key가 없거나, 중복 저장
            # v가 루트인 경우
    def isBalanced(self, v):
        if v == None: return "### AVL.isBalanced : The Node not found ###"
        L = -1 if v.left == None else v.left.height
        R = -1 if v.right == None else v.right.height
        return abs(L-R) <= 1
    
    # BST.insert() 호출 => revalance()
    def insertAVL(self, key):
        # 상속받은 BST 클래스의 insert 호출 후 반환 값을 담는다.
        v = self.insert(key)

        # 중복된 key 혹은 key값 자체가 None인 경우
        if v is None : return None 
        # v가 루트노드인 경우
        if v == self.root : return v
        # 루트의 자손인 경우 (아직 3대를 이루지 못함)
        if v.parent == self.root : return v
        # 여기에 걸러지지 않았다면, 삽입된 v는 3대 이상임 따라서 x, y, z 존재

        x, y, z = v, v.parent, None
        # 노드가 삽입된 h=0인 자리 x에서부터 한 칸씩 올라가면서, z에서 균형이 맞는지 검사한다.
        # 균형이 안맞는 노드(회전의 중심 z)를 찾기 위해 위로 탐색
        while y:
            z = y.parent
            # z의 존재 확인 : y의 부모가 있는 경우(즉 삽입된 x 위로 두 세대 더 존재하는 경우)
            if z:
                # z의 left와 righte를 비교하여 그 값이 1, 0, -1인지 확인
                if self.isBalanced(z):
                    # 균형이 맞는다면 한칸씩 위로 올라감
                    x = y
                    y = z
                    # z는 y가 존재하면 자동으로 y.parent
                #불균형 포인트 찾은 뒤엔 while문 탈출
                else:
                    break
            # root까지 조사해봤으나, 균형을 이루는 상태니, 따로 rebalance 불필요
            else: return v
        # x, y, z 찾음 (3세대 이상이며, 불균형한 지점을 찾음)
        w = self.rebalance(x, y, z)
        # root가 바뀐 경우 갱신
        if w.parent is None:
            self.root = w
        # 무사히 삽입 끝난 뒤엔, 처음 insert한 값 반환
        # 루트, 혹은 루트의 자손이 아닌 경우에만 여기까지 도달
        return v
        # rebalance -> rotateRight/left의 마지막에 언제나 updateHeight()호출 하므로, 높으는 최신 상태 유지   

    def deleteAVL(self, key):
        # 삭제 후 균형이 깨질 가능성이 있는 가장 깊은 노드 반환
        z = self.deleteByMerging(key)
        # root가 바뀔 경우 갱신하기 위해 
        w = self.root
        # z가 root가 될 때까지 계속 올라간다
        while z is not None:
            # 균형이 잘 맞는다면 계속 위로 올라가
            if not self.isBalanced(z):
                # 균형 깨진 노드 기준으로 더 깊은 곳을 탐색하기 위해 양쪽의 서브트리 높이 비교
                Lz = -1 if z.left == None else z.left.height
                Rz = -1 if z.right == None else z.right.height
                # 오른쪽이 더 깊다면
                if Lz <= Rz:
                    # z.left가 none일 수도 있음
                    y = z.right
                else:
                    y = z.left
                if y:
                    Ly = -1 if y.left == None else y.left.height
                    Ry = -1 if y.right == None else y.right.height
                    if Ly <= Ry:
                        x = y.right
                    else:
                        x = y.left
                    z = self.rebalance(x, y, z)
            w = z
            z = z.parent
        self.root = w
```

```python
"""
[test] rebalance
  1) 아직 균형이 맞춰지지 않은 이진 탐색 트리(BST) 생성 - BST.insert() 사용
  2) isBalanced()로 균형이 맞지 않는 부분을 정상적으로 체크하는 지 확인
  3) AVL.rebalance() 작동 확인
"""

# 1) 아직 균형이 안 맞춰진 이진 탐색 트리 생성
A = AVL(20)
Keys = [31, 25, 23, 26, 33, 22]
Nodes = {key : A.insert(key) for key in Keys if key is not None}
A.printInfo("[step1] Creating unbalanced BST tree (z = 31)")

# 2) isBalanced()로 균형이 맞지 않는 부분을 정상적으로 체크하는 지 확인
print("*** [step2] IsBalanced? ***")
for key in Nodes.values():
  print(f"  - Node({key})  =>  {A.isBalanced(key)}")
print()

# 3) AVL.rebalance() 작동 확인
# z(부모) - y(자식) - x(손자)
A.rebalance(Nodes[23], Nodes[25], Nodes[31])
A.printInfo("[step3] After Rebalance(key=31)")
for key in Nodes.values():
  print(f"  - Node({key})  =>  {A.isBalanced(key)}")
```
