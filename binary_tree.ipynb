```python
class Node:
    def __init__(self, key):
        self.key = key
        self.parent = None
        self.left = None
        self.right = None
        
    def __str__(self):
        return str(self.key)
    
    # preorder 순회를 default
    def __iter__(self):
        if self != None:
            yield self.key
            if self.left != None:
                for leftSubTree in self.left:
                    yield leftSubTree
            if self.right != None: 
                for rightSubTree in self.right:
                    yield rightSubTree
```

```python
class BinaryTree:
    def __init__(self, key=None):
        self.root = Node(key)
        self.size = 1
        
    def __str__(self):
        return str(" ").join(str(i) for i in self.root)
    
    def preorder(self, now):
        if now != None:
            print(now, end=" ")
            if now.left != None:
                self.preorder(now.left)
            if now.right != None:
                self.preorder(now.right)
                
    def inorder(self, now):
        if now != None:
            if now.left != None:
                self.inorder(now.left)
            print(now, end=" ")
            if now.right != None:
                self.inorder(now.right)

        
    def postorder(self, now):
        if now != None:
            if now.left != None:
                self.postorder(now.left)
            if now.right != None:
                self.postorder(now.right)
            print(now, end=" ")
        
    def insertLeft(self, Node, NewNode):
        Node.left = NewNode
        NewNode.parent = Node
        self.size += 1
        
    def insertRight(self, Node, NewNode):
        Node.right = NewNode
        NewNode.parent = Node
        self.size += 1
```
