                       한방향 연결 리스트, Single Linked List Class

                    [head, size] ---->  []->[]->[]->[]->[]->None

    노드가 링크에 기차처럼 연결된 구조로, 링크를 따라 원하는 노드에 접근할 수 있다. 
    index를 통해 요소에 접근 하는 게 아니라, head노드로부터 link를 따라 원하는 요소를 찾을 수 있다.
    따라서 각 요소를 표현하는 node클래스와, 전체 정보를 담고 있고, 연산을 제공하는 Single Linked List를 각각 정의하자. 

1. Node Class 정의

                                            [key, next]

    규칙 : 
        전체 구조가 아닌, 각각의 노드 개인(?)을 정의한다. 
        노드에 저장할 값을 대표하는 key와 다음 요소의 주소를 담을 next를 맴버 변수로 갖는다.
        필요에 따라 value를 추가할 수 있다.
    특수 메서드 : __str__ 

1.1. Node Class 구현
        생성자 호출 시, key을 인자로 받아 초기화, next 초기값은 None => 맨 끝 노드 따로 지정해주지 않아도 괜찮음
        __str__호출 시, key 값만 반환



2. Single Linked List Class 정의 #전체 관리 클래스

                                    [head, size] ---> []-[]-[]-...-[]

    규칙 : 연결된 노드들의 정보를 표시하고, 메서드를 제공하는 클래스로, 가장 앞의 노드인 head와 전체 노드의 개수 size를 맴버 변수로 갖는다.
    연산 : 탐색(search), 삽입(pushFront,pushBack), 삭제(popFront, popBack, remove)
    특수 메서드 : __iter__, __str__, __len__

2.1. Single Linked List Class 구현
    생성자 
        node 인스턴스를 전체 관리하는 역할로 head와 size를 맴버 변수로 갖는다.

    메서드
    pushFront 
        1) 노드 생성 
        2) 생성된 노드의 next를 기존의 head 노드로 지정
        3) 관리 객체의 head와 size정보도 갱신                          
        """관리 객체 : Single Linked List Class"""
        # 만약 첫 번째 삽입이라면? 
            생성된 노드를 head 노드로 지정하기, size 증가 (어차피 다음 링크는 None이라서 지정 필요 x)
    pushBack
        # 이미 하나 이상의 노드가 존재하는 경우 
            1) 노드 생성
            2) 마지막 노드를 탐색하여 반환
            3) 위에서 찾은 노드의 next를 생성된 노드로 지정
            4) 관리 객체의 size정보 증가 (head 변화 없음)
        # 만약 첫 번째 삽입이라면?
            생성된 노드를 head로 지정, size증가
    popFront
        # 하나 이상의 요소가 있는 경우 (노드가 하나뿐이었다면, 자동으로 다음 요소가 None)
            1) 관리 객체의 head를 기존 head의 next로 지정
            2) size 감소
            3) 기존 head 노드 반환
        # 만약 텅 빈 연결리스트라면?
            비어있다고 말해주기 
    popBack
        # 노드가 하나만 있는 연결리스트에서 요소 삭제 시
            1) 관리 객체의 head를 None으로 지정
            2) size 감소
            3) 기존 head 노드 반환
        # 노드가 둘 이상 있는 연결리스트의 요소 삭제 시
            1) 마지막 노드와, 마지막 이전 노드 탐색
            2) 마지막 이전 노드의 next를 None으로 지정
            3) size 감소 (head 변화 x)
            4) 기존 마지막 노드 반환
        # 만약 텅 빈 연결리스트라면?
            비어있다고 말해주기 
    remove(v) #node의 주소를 이용해 접근하는 게 아니라, key값을 통해 조회하여 삭제할 수도 있음
        # head 노드를 지우는 경우
            1) 관리 객체의 head를 기존 head의 next로 지정
            2) size 감소
            3) 기존 head 반환
        #head가 아닌 노드를 지우는 경우
            1) 해당 요소 A와 그 이전 요소 B를 동시에 탐색 후 반환
            2) 이전 요소 B의 next를 None으로 지정
            3) size 감소
            4) 해당 요소 A 반환 
        # 지우려는 연결리스트가 텅 비어있거나, 애초에 None을 지워달라고 요청한 경우
            무시한다

    특수 메서드
    __iter__ 
        1) head값부터 시작해서 tail에 도달할 때까지, 어디선가 사용된 for 문에 yield를 통해 전달 
        2) 전달 후 현재 값을 다음 노드로 지정
    __str__
        1) 위에서 구현한 for문을 순회하며 각 노드 사이에 ->를 추가하여 연결
    __len__
        관리 객체의 size 값 반환

    search(key)
        1) 제너레이터를 통해 구현된 for문을 이용하여 해당 key 값을 만나면 반환
