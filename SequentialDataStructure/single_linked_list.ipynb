한방향 연결 리스트, Single Linked List Class

                                [head, size] ---->  []->[]->[]->[]->[]->None

    노드가 링크에 기차처럼 연결된 구조로, 링크를 따라 원하는 노드에 접근할 수 있다. 
    index를 통해 요소에 접근 하는 게 아니라, head노드로부터 link를 따라 원하는 요소를 찾을 수 있다.
    따라서 각 요소를 표현하는 node클래스와, 전체 정보를 담고 있고, 연산을 제공하는 Single Linked List를 각각 정의하자.

        필요에 따라 value를 추가할 수 있다.
    특수 메서드 : __str__

2. Single Linked List Class 정의 #전체 관리 클래스

                                    [head, size] ---> []-[]-[]-...-[]

    규칙 : 연결된 노드들의 정보를 표시하고, 메서드를 제공하는 클래스로, 가장 앞의 노드인 head와 전체 노드의 개수 size를 맴버 변수로 갖는다.
    연산 : 탐색(search), 삽입(pushFront,pushBack), 삭제(popFront, popBack, remove)
    특수 메서드 : __iter__, __str__, __len__

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.next = None
        
    def __str__(self) -> str:
        return str(self.key)
```

```python
from itertools import count


class singleLinkedList:
    def __init__(self):
        self.head = None
        self.size = 0

    # pushFront [v] + [v]->[]->[]->0
    def pushFront(self, key):
        newNode = Node(key)
        newNode.next = self.head #이미 요소가 있거나 없거나 잘 작동
        # 위에서 self의 요소가 Node 인스턴스를 참조하기 때문에, 
        # 현재 클래스에서 선언한 맴버변수가 Node 클래스의 맴버변수를 참조할 수 있음
        self.head = newNode 
        self.size += 1
        
    # pushBack      []->[]->[v]->0 + [v]
    def pushBack(self, key):
        newNode = Node(key)
        self.size += 1
        #빈노드면? head만 갱신
        if self.size == 0:
            self.head = newNode
        #비어있지 않다면? 마지막 노드 찾아서 그녀석의 next를 new로 지정
        else:
            tail = self.head
            while tail.next != None:
                tail = tail.next
            tail.next = newNode
            
    def popFront(self)->(Node):
        #비어있다면? 무리무리 
        if self.size == 0:
            print("Empty")
            return None
        #비어있지 않다면? 무조건 헤드노드 갱신
        else:
            v = self.head 
            self.head = self.head.next
            self.size -= 1
            return v
        
    def popBack(self)->(Node):
        #비어있다면? 이무것도 안해
        if self.size == 0:
            print("Empty")
            return None
        #노드가 딱 하나만 들어있다면? 삭제 후 헤드노드 갱신
        elif self.size == 1:
            v = self.head
            self.head = None
            self.size -= 1
            return v
        #노드가 둘 이상 들어있다면? 마지막 값과 이전 값 찾아서 리턴 삭제 후 갱신 불필요
        else:
            preTail, tail = self.head, self.head.next
            while tail.next != None:
                tail = tail.next
                preTail = preTail.next
            preTail.next = None
            self.size -= 1
            return tail

    def __iter__(self):
        v = self.head
        while v != None:
            yield v
            v = v.next
            
    def __str__(self) -> str:
        return "->".join(str(v.key) for v in self)
    
    def __len__(self):
        size = 0
        for v in self:
            size += 1
        return size
```

```python
L = singleLinkedList() 

print("***PushFront***")
L.pushFront(1)
L.pushFront(2)

print(L)
print("head is", L.head)
print("size is", L.size)
print("real size is", len(L))


```

```python
print("***PushBack***")
L.pushBack(1)
L.pushBack(2)

print(L)
print("head is", L.head)
print("size is", L.size)
print("real size is", len(L))
```

```python
print("Before popFront", L)

for i in L:
    print()
    print(L.popFront(), L)
    print("head is", L.head)
    print("size is", L.size)
    print("real size is", len(L))
```

```python
print("Before popBack", L)

for i in L:
    print()
    print(L.popBack(), L)
    print("head is", L.head)
    print("size is", L.size)
    print("real size is", len(L))
```
