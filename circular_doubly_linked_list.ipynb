        근데 a,b와 x가 다른 LinkedList 인스턴스에 속해있는 경우는 size를 어떻게 다뤄야할까?

```python
# Node Class
"""
__init__
    key를 인자로 받아 초화하고, next, prev는 자기 자신으로 초기화한다. (필요에 따라 value를 가질 수 있다) 
__str__
    key값을 string으로 반환한다.
"""
class Node:
    def __init__(self, key):
        self.key = key
        self.next = self
        self.prev = self
    def __str__(self) -> str:
        if self.key is None:
                return "[]"
        else:
            return str(self.key)
```

```python
# Circular Doubly Linked List

#생성하는 경우를 제외하고, key에 None이 오면 모두 예외로 처리
# head를 구분하는 방식
class CirularDoublyLinkedList:
# Special Method
    def __init__(self):
        self.head = Node(None)
    # head 노드를 읽을 필요는 없지만, 구현 과정에서 잘 작동하는지 알아보기 위해, 시작과 끝에 head가 보이게 작성
    def __iter__(self):
        v = self.head
        while v.next != self.head:
            yield v
            v = v.next
        yield v
        yield v.next
    # Node의 str 메서드를 직접 호출, 좋은 방법인지는 잘 모르겠다. 의존성이 증가하니까..!
    def __str__(self) -> str:
        return "-".join(Node.__str__(v) for v in self)
    # 실제 리스트를 순회하며 개수를 카운트, 개수에 head를 포함하지 않는다.
    def __len__(self):
        realSize = 0
        for v in self:
            realSize += 1
        return realSize - 2
    # List의 전체 구조, head, isEmpty, size 정보 표시
    def printInfo(self):
        isEmpty = "Yes" if self.isEmpty() else "No"
        print("    =>    info")
        print("          - List structure  | ", self)
        print("          - head:", self.head, " | isEmpty:", isEmpty, " | count(node):", len(self))
        print()
        
# Instance Method
# 기반 연산
    def splice(self, a, b, x):
        # abx의 존재 여부 검사
        if a is None and b is None and x is None:
            return
        # 검사 사항  
        # a와 b가 동일하면 빠져나온다
        # a 다음을 따라가다보면 b를 만나야한다
        # 근데 검사 행위 자체가, 경우에 따라서 무한 루프를 돌게 하네 그럼 while문이 아니라 for each로 한바퀴만 돌아야겠네

        testIndex = 0
        aIndex= 0
        bIndex=0
        xIndex = 0 # 새로운 예외 때문에 추가
        # 문제 1 : 제자리에 복붙하면 x 다음에 a가 와서, xn이 a인데 그럼 x-a..b-ab-ab-ab.. 무한 루프에 빠짐
        # 문제 2 : a와 b사이에 x가 존재할 때, ab를 자르는 건 가능인데, 이때 붙여넣을 곳이 사라져서 문제 => 오직 잘라내기만 발생 (삭제 기능에 활용 가능)
            # 구현 방식에 따라서 자른 뒤에 붙일 게 아니라, 이를 반대로 수행하면 (일정 노드를 떼서 붙이기 가능할듯ㅋ)
        # 문제 3 : a와 x가 같을 때, b와 x가 같을 때, 모두 잘라내어짐 
        # 필요에 따라서 문제 123을 활용하면 될듯 일단은 막아둘게요 ㅅㄱ
        # 만약 a와 b가 같으면 리스트에 존재하면 옮기고, 없으면 생성하잖아. 그니까 아무 문제 없는 거 따라서 a와 b가 다른 경우만 비교

        # a와 b가 다른 경우에 모든 노드 검사해서 둘 다 있는지 검사
        for node in self:
            # 몇 번째에 방문했는 지 확인
            testIndex  += 1
            if node == a:
                aIndex = testIndex 
            if node == b:
                bIndex = testIndex
            if node == x:
                xIndex = testIndex
                
        if a == b and aIndex - xIndex != 1:
            pass
        # 332 에러 제자리 복붙인데 a와 b가 같은 경우
        elif a==b and aIndex - xIndex == 1:
            return
        else:
            # //// for 문을 한바퀴 돌면서 모든 노드를 검사. aIndex와 bIndex에 발견한 순서 있음
            #a와 b중에 하나 또는 모두를 발견 못한 경우 => 메시지 반환 후 함수 종료
            # !(aIndex || bIndex)로 대체 가능
            if aIndex == 0 or bIndex == 0:
                #print("### splice : 잘라내기 할 요소를 찾을 수 없습니다. a나 b가 리스트에 존재하는지 확인해주세요. ###")
                return
            # x 인덱스가 다른 리스트 상에 존재한다면 pass
            elif xIndex == 0:
                pass
            # 모든 게 0이 아니라면, 이제 x가 ab사이에 있는지, x다음에 ab 다음에 오는지 확인해야함, 이외의 경우만 정상
            else:

            # x가 a와 b 사이에 있거나 자리가 바뀌는 바람에 b와 a 사이에 있는 경우 예) 2,7,5 또는 7,2, 5
                if aIndex <= xIndex <= bIndex or bIndex <= xIndex <= aIndex:
                    #print("### splice : a와 b 사이 닫힌 구간에 x가 존재하는 경우는 뭔가 곤란합니다. 아무런 일도 일어나지 않습니다. 삭제나 잘라내기를 구현할 때 사용할 예정")
                    return
                # x 다음에 a가 등장하는 경우(자르고 원래 자리에 붙이기) 또는 ab순서가 바꼈는데, x다음에 b가 등장하는 경우도 동일 
                if aIndex - xIndex == 1:
                    #print("잘랐다가 원래 자리에 붙이는 경우니까 아무것도 하지 않고 원본 반환")
                    return
                # x가 리스트 내에 존재하지만, ab사이에 있지도 않고, 제자리에 복붙하는 것도 아닌 경우 중에서.. ab의 순서 검사
                else:
                    # 정상적인 순서인 경우 검사 탈출
                    if aIndex < bIndex:
                        pass
                    # a와 b의 자리가 바뀐 경우 원상 복귀 후 계속 진행
                    # 532 에러-> 제거 
                    elif aIndex > bIndex:
                        # 자리 바꾼 뒤에 ab 했을 때, x다음 ab가 올 수도 있으니까 예시: (5, 3, 2) 제자리에 순서 바꿔서 붙이는 경우
                        if  bIndex-xIndex == 1:
                            #print("### splice : 제자리에 복사 붙여넣기를 할 수 없고, 심지어 순서도 바꼈습니다. ###")
                            return
                        # 자리를 원래대로 뒤집었는데, 문제가 없는 경우
                        else:
                            #print("### splice : a와 b 노드의 자리가 바꼈습니다. 두 노드의 순서를 자동으로 바꿔서 계산합니다.  ###")
                            a, b = b, a
                            pass
            
            # key값이 서로 다른 주소를 갖는 노드를 x뒤에 삽입할 때? 이건 앞 조건에서 걸리지 않나
        #한 바퀴를 돌면서 a가 나오면 체크 b가 나오면 체크 x가 나오면 체크
        #특히 a가 먼저 나오면 굳, 그리고 b 탐색 
        # 어디까지 허용할까. a와 b가 같다면 둘 다 해당 리스트에 없어도 괜찮음
        # a와 b의 자리가 바꼈다면, 메시지 반환하거나 혹은 둘의 자리 교환
        # x가 해당 노드에 존재하지 않으면? 그건 상관 없을듯 => 다른 리스트에 붙일 수 있으니까
        # 그럼 내가 확인할 사항 : for 문 속에서 a와 b가 등장하는 지, a가 먼저 나오고 b가 나왔는지, b가 먼저나오고 a가 나왔는지, a나 b 중에 하나가 없는지, 둘 다 없는지=>이떄 a와 b가 같다면 허용
        # if a 등장: a_b_ab_ba_ax_bx_xx    
        # a -> ab : break
        # b-> ba : change and break
            
        
        # 검사 성공
        ap = a.prev
        bn = b.next
        xn = x.next
        # Cut       [ap]<->[bn]
        ap.next = bn
        bn.prev = ap
        # Paste     [x]<->[a]..[b]<->[xn]
        a.prev = x
        x.next = a
        b.next = xn
        xn.prev = b
        # 게다가 서로 다른 관리 객체에 대해 연산 진행 시 어떻게 해야할지 햇갈림 
        # size 문제인데 아예 맴버변수에서 size를 제외하고, len을 통해서만 호출하게 할 지, 또는 size변수 자체가 늘 len함수를 호출해서 업데이트하게 할 지 고민... 너무 비효율적

# 탐색 연산 
    def isEmpty(self)->(bool):
        return (True if self.head.next == self.head and self.head.prev == self.head else False)
    
    def search(self, key) ->(Node): 
        # 비어있는 경우
        if self.isEmpty(): 
            #print("### search : 빈 리스트입니다 ###")
            return 
        # None을 검색하는 경우 (사실 이건 head인뎅 ㅋㅋ)
        if key is None:
            #print("### search : 유효하지 않은 값을 입력했습니다 ###")
            return 
        nowNode = self.head.next
        while nowNode != self.head:
            if nowNode.key == key:
                return nowNode
            nowNode = nowNode.next
        #print("### search : 일치하는 값을 찾지 못했습니다 ###")
    # last()  -> (Node)
    # first() -> (Node)
    
# 삽입 연산과 이동 연산
    def pushFront(self, key):
        a = Node(key)
        self.splice(a,a, self.head)
    def pushBack(self, key):
        a = Node(key)
        self.splice(a,a, self.head.prev)
    def insertBefore(self, key, x):
        a = Node(key)
        self.splice(a, a, x.prev)
    def insertAfter(self, key, x):
        a = Node(key)
        self.splice(a, a, x)
    def moveBefore(self, a, x):
        self.splice(a, a, x.prev)
    def moveAfter(self, a, x):
        self.splice(a, a, x)

    # Splice 없는 버전
    def pushFront2(self, key):
        # None을 삽입하려는 경우 : 안돼 중복된 키는 어떡하지 일단 허용하되, 이후 다른 자료구조를 통해 부족한 부분을 보완해보자. 해시와 매핑한다거나
        if key is None:
            print("### pushFront : 값을 입력하세요 ###")
            return
        # 첫번쨰로 삽입되는 노드인 경우 :  [head] <-> [newNode] <-> [head]
        newNode = Node(key)
        if self.isEmpty():
            self.head.next = newNode
            self.head.prev = newNode
            newNode.next = self.head
            newNode.prev = self.head
        # 아닌 경우 :         [head] <-> [newNode] <-> [firstNode] ...
        else:
            firstNode = self.head.next #기존의 첫번째 노드를 참조하는 변수 생성
            self.head.next = newNode
            newNode.prev = self.head
            newNode.next = firstNode
            firstNode.prev = newNode
    def pushBack2(self, key):
        # None을 삽입하려는 경우 : None
        if key is None:
            print("### pushBack : 값을 입력하세요 ###")
            return
        # 리스트가 비어있는 경우 :  pushFront 재사용
        newNode = Node(key)
        if self.isEmpty():
            self.head.next = newNode
            self.head.prev = newNode
            newNode.next = self.head
            newNode.prev = self.head
        # 아닌 경우 :         [head] <-> [tailNode] <-> [newNode] <-> [head]
        else:
            tailNode = self.head.prev #기존의 마지막 노드를 참조하는 변수 생성
            self.head.prev = newNode
            newNode.next = self.head
            newNode.prev = tailNode
            tailNode.next = newNode

    # 삭제 연산
    # remove(x) -> (node)     x 노드를 찾아서 삭제 or delete(key) 가 편할듯?
    # popFront() -> (node)    맨 앞의 값 삭제
    # popBack() -> (node)     맨 뒤의 값 삭제
    
    #   [xp]<->[xn]
    def remove(self, x)->(Node):
        if x is None: return 
        if self.isEmpty(): return
        xp = x.prev
        xn = x.next
        xp.next = xn
        xn.prev = xp
        return x
```
