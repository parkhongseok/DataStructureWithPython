<h1>
Banary Tree & Heap ?
</h1>

Heap은 엉덩이가 아니다. 이진 트리다.
표현 방식과, 특정 조건 두가지를 만족한다면 힙이라고 부른다. 우선 표현 방식에 대해 알아보자.

각 요소들이 일대다 관계를 갖는 경우, 그래프라고 부르며, 특히 부모와 자식의 관계만으로 구성되어있고, 순환하지 않는 경우를 트리라고 한다. 
자식을 최대 둘까지 갖는 경우를 이진트리라고 한다.
이진트리를 코드상으로 표현할 수 있는 방법 중 세 가지를 소개한다. 

방법 1. 리스트 상에 레벨별로 나열하기 
    모든 노드가 있다고 생각하고, 각 레벨별로 모든 요소를 차례대로 표현하는 방식이다.
    빈 노드의 경우 None으로 두고, 가장 왼쪽 노드부터 오른쪽으로 적는다.
    빈 노드가 많으면 메모리 낭비가 심하다.

방법 2. 리스트에 재귀적으로 표현하기
    [자신, [왼쪽 서브트리], [오른쪽 서브트리]] 와 같은 방식을 재귀적으로 표현하는 방식이다. 

방법 3. 노드 클래스를 정의해서 이를 연결리스트와 비슷한 방식으로 연결하는 방식이다.
노드 클래스는 (key, value), 부모링크, 왼쪽자식링크, 오른쪽자식링크 이렇게 총 다섯개의 요소를 적어도 갖는다. 


위에서 표현 방식 세가지를 알아봤고, 아래에선 힙이 만족해야하는 성질 두 가지를 소개한다.

모양 성질 : 리프 노드를 제외한 모든 노드가 완전히 차있다. 마지막 레벨에선 왼쪽부터 값이 차있다.
힙 성질 : 루트를 제외한 모든 노드의 값은 자신의 부모노드보다 크면 안된다.

<h3>Heap : 힙은 이진트리 중에서도 다음의 두 가지 성질을 모두 만족하는 트리를 [방법1]인 리스트 형식으로 표현한 자료구조이다.</h3>

```python
class MaxHeap:
    def __init__(self, L=[]) -> None:
        self.H = L                   #모양성질 만족
        self.makeHeap()            #힙성질 만족
        
    def __str__(self):
        return str(self.H)
    
    def __len__(self):
        return len(self.H)
    
    def heapifyDown(self, now, size):                   #makeHeap과 heapSort에 사용됨
        # n= 힙의 전체 노드 개수 [heapsort를 위해 필요함]
        # 트리 입장에서 아래로 내려가며 A[i]가 힙성질을 만족하도록 재배치
        # 트리의 루트에서 내려가며 나의 첫째 자식 인덱스가 전체 노드 수와 같다 => 내 다음인덱스부터 오는 모든 노드는 자식을 가질 수 없다 == 루트노드이다
        # 즉, 리프노드에 대한 조건이다. 첫째 자식이 있으려면 적어도, 그녀석의 인덱스가 실제 마지막 인덱스는 되어야 한다는 것이다. 그렇지 않고 마지막을 넘어가면 전부 자식이 없으니 리프인거다.
        # 마지막 인덱스는 전체 요소의 크기에서 하나를 제외한 것이다. 오른쪽 요소가 존재
        while 2*now+1 < size : # L이 n인 순간부터 리프노드다. 리프노드가 되면 while문 탈출
            Left, Right = 2*now+1, 2*now+2
            # 리프노드가 아니니까, 왼쪽 자식은 적어도 무조건 갖고 있으니까 단순히 비교 => 오른쪽 자식은 없을 수도 있으니 존재여부 따져봐야함
            if self.H[now] < self.H[Left]: #자신과 왼쪽자식 비교
                maxNodeIndex = Left
            else:
                maxNodeIndex = now
            # 지금 노드가 Right를 가지려면 존재하려면, 내 오른자식 인덱스가 사이즈와 같아지는 순간부터는 존재자체가 불가능. 이보다는 작아야함 
            if Right < size and self.H[maxNodeIndex] < self.H[Right]:
                maxNodeIndex = Right
            # 이제 최대값을 찾았으니까, 만약 그럼 현재 내 자리에 최고를 앉히자.
            if now != maxNodeIndex:
                # 자리를 바꾸고
                self.H[now], self.H[maxNodeIndex] = self.H[maxNodeIndex], self.H[now]
                # 최대 노드가 있던 자식의 자리로 내 자리를 옮겨간다.
                now = maxNodeIndex
            # 혹은 이미 내 자리가 최대라면, 현재 나와 내 자식노드 총 셋은 힙 성질을 만족하니까 멈추자.
            else:
                break
    # heapifyDown을 List기준 절반지점부터 root를 향한다. <<< 왜 이런 순서로 검사하는 걸까. 그냥 root부터 아래로 내려가며(list로 따지자면 index 0부터 쭉 보면 안되나?)
    # 안된다. 귀류법으로 살펴본다면, 만약 그런 방법으로 정렬했을 때, root노드에 최대값이 오는가? 이다.
    # 만약 leaf노드에 최대값이 있었다고 한다면, 이 연산을 한번 수행했을 때, 고작 한 레벨 위로 올라갔을 거다.
    # 비유하자면 최강자를 찾기 위해서 지역대회를 뚫고 올라온 자들을 본선에서 붙여야지, 
    # 냅따 그냥 서울사는 사람 중에 1등 뽑고나서, 서울 2, 3등과 호남지방과 비교해서 다시 2, 3등 결정하고, 이지랄 하면 또 1등과 2, 3등 비교하고, 
    # 이걸 전체 트리의 높이만큼은 반복해야한다. 즉, 훨씬 노력이 많이 필요하게 된다.
    def makeHeap(self):
        size = len(self.H)
        # for now in range(size-1, -1, -1):의 범위를 for now in range(size//2, -1, -1):로 바꿔도 문제가 없다.
        # 왜냐하면 size는 리스트의 마지막 인덱스보다 하나 큰데, 전체 노드의 개수가 짝수라면 리프와 리프가 아닌 노드의 개수가 완전히 똑같고, 홀수라면 리프의 개수가 하나 더 많기 때문에
        # 절반만큼의 인덱스, 또는 절반+1칸만큼까지만 검사해도 충분하다. 
        # leaf가 나오면 heapifyDown에서 Left < size 조건에서 걸러지기 때문에, 검사 범위가 leaf를 포함해도 상관 없다. 참고로 //연산은 나눈 뒤 정수 부분만을 반환한다
        for now in range(size//2, -1, -1):
            self.heapifyDown(now, size)
            
    def heapSort(self):
        size = len(self.H)
        self.makeHeap()
        while size != 0:
            self.H[0], self.H[size -1] = self.H[size - 1], self.H[0]
            size -= 1
            self.heapifyDown(0, size)
    
    # 마지막 인덱스를 기준으로
    def heapifyUp(self, now):
        # 4//2와 5//2 연산 모두 그 결과가 2다. 따라서, 현재 노드가 L인지 R인지 구분할 필요 없이 부모노드를 구할 수 있다.
        # root에 도착하거나, 부모가 더 크다면 while을 탈출한다.
        while now != 0 and self.H[(now-1)//2] < self.H[now]:
            Parent = (now-1)//2
            # 부모보다 자식이 큰 경우에 현재 반복문에 들어오니까, 값과 인덱스의 자리를 모두 바꾼다. 
            self.H[Parent], self.H[now] = self.H[now], self.H[Parent]
            now = Parent

    # append로 삽입 후 사실 위에서 만들어둔 makeHeap연산을 사용해도 된다. 하지만 매우 비효율적이다. 왜냐면 삽입된 녀석만 빼면 모두 heap성질을 만족하도록 정렬되어있는데, 
    # 굳이 다시 루트부터 아래로 내려가며 leaf가 아닌 모든 노드를 검사할 필요가 없다.
    # 만약 삽입된 값이 최대값이라고 하더라도, 자신의 부모와 자리를 차근차근 바꿔가며, root를 기준으로 내가 속한 subTree만 바꿔도 충분할 것이다.
    def insert(self, key):
        self.H.append(key)
        self.heapifyUp(len(self.H)-1)
        
    def deleteMax(self):
        if len(self.H) == 0:
            return None
        # 최대 값을 따로 보관
        max = self.H[0]
        # 리스트의 양 끝값 스왑
        self.H[0], self.H[len(self.H)-1] = self.H[len(self.H)-1], self.H[0]
        # 마지막 값(최대) 제거
        self.H.pop()
        # 얼떨결에 첫번째에 오게 된 노드를 기준으로 다시 정렬 진행
        self.heapifyDown(0, len(self.H))
        return max

    def printHeapStructure(self, index=0, prefix="", is_left=True, is_root=True):
        heap = self.H

        if index < len(heap):
            node_value = heap[index]

            if is_root:
                print(f"──── {node_value} (idx={index})")
            else:
                print(prefix + (" ├── " if is_left else " └── ") + f"{node_value} (idx={index})")

            left_child_index = 2 * index + 1
            right_child_index = 2 * index + 2

            if left_child_index < len(heap) or right_child_index < len(heap):
                if left_child_index < len(heap):
                    self.printHeapStructure(left_child_index, prefix + (" │    " if not is_root and is_left else "      "), True, False)
                else:
                    print(prefix + (" │    " if not is_root and is_left else "      ") + " ├── None")

                if right_child_index < len(heap):
                    self.printHeapStructure(right_child_index, prefix + (" │    " if not is_root and is_left else "      "), False, False)
                else:
                    print(prefix + (" │    " if is_left else "      ") + " └── None")
                    
    def getHeight(self)->int:
        from math import floor, log2
        return 0 if not self.H else floor(log2(len(self.H)))

    def info(self, msg=""):
        print("\n" + str(msg))
        self.printHeapStructure()
        print("=>  info")
        print("    - Heap Structure  | ",  self.H)
        print("    - Number of Nodes | ", self.__len__())
        print("    - Tree Height     | ", self.getHeight(), "\n")
```
