<h1> 한방향 연결 리스트, Single Linked List Class?</h1>
                                [head, size] ---->  []->[]->[]->[]->[]->None

    노드가 링크에 기차처럼 연결된 구조로, 링크를 따라 원하는 노드에 접근할 수 있다. 
    index를 통해 요소에 접근 하는 게 아니라, head노드로부터 link를 따라 원하는 요소를 찾을 수 있다.
    따라서 각 요소를 표현하는 node클래스와, 전체 정보를 담고 있고, 연산을 제공하는 Single Linked List를 각각 정의하자. 

<h1>
1.Node Class 
</h1>
- Attributes
    `key`
    `next`

- Rules
    전체 구조가 아닌, 각각의 노드를 정의한다. 첫번째 노드를 head, 끝을 tail이라고 하자.
    노드에 저장할 값을 대표하는 key와 다음 요소의 주소를 담을 next를 맴버 변수로 갖는다.
    필요에 따라 value를 추가할 수 있다.

- Methods
    Special Methods 
    __init__        => key을 인자로 받아 초기화, next 초기값은 None 
                    => 맨 끝 노드 따로 지정해주지 않아도 괜찮음
    __str__         => key 값을 String 타입으로 반환



<h1>
2.Single Linked List Class 
</h1>
- Attributes
    `head`
    `size`

- Rules
    연결된 노드들의 정보를 표시하고, 메서드를 제공하는 클래스로, 가장 앞의 노드인 head와 전체 노드의 개수 size를 맴버 변수로 갖는다.
    
- Methods
    Special Methods
    `__init__`      => node 인스턴스를 전체 관리하는 역할로 head와 size를 맴버 변수로 갖는다.
    `__iter__ `     => head값부터 시작해서 tail에 도달할 때까지, 어디선가 사용된 for 문에 yield를 통해 전달 
                    => 전달 후 현재 값을 다음 노드로 지정
    `__str__`       => 위에서 구현한 for문을 순회하며 각 노드 사이에 "->"를 추가하여 연결
    `__len__`       => 인스턴스 변수의 size 값 반환 (하지만 여기선, size값이 올바르게 변하는지 비교를 위해 직접 카운트)

    Instance Methods
    `pushFront `    => 새로운 노드 객체 생성  
                    => 생성된 노드의 next를 기존의 head 노드로 지정
                    => 관리 객체의 head와 size정보도 갱신                          
                        # 만약 첫 번째 삽입이라면?
                            => 생성된 노드를 head 노드로 지정하기, size 증가 
                                (어차피 다음 링크는 None이라서 지정 필요 x)

    `pushBack`      => 새로운 노드 객체 생성
                    => size 증가
                    if 이미 하나 이상의 노드가 존재하는 경우 
                        => 마지막 노드를 탐색
                        => 위에서 찾은 노드의 next를 생성된 노드로 지정
                    else 만약 첫 번째 삽입이라면?
                        => 생성된 노드를 head로 지정

    `popFront`      if 하나 이상의 요소가 있는 경우 
                        (노드가 딱 하나뿐이었다면, 자동으로 다음 요소가 None)
                        => 관리 객체의 head를 기존 head의 next로 지정
                        => size 감소
                        => 기존 head 노드 반환
                    else 만약 텅 빈 연결리스트라면?
                        => 비어있다고 말해주기 

    `popBack`       if 노드가 하나만 있는 연결리스트에서 요소 삭제 시
                        => 관리 객체의 head를 None으로 지정
                        => size 감소
                        => 기존 head 노드 반환
                    elif 노드가 둘 이상 있는 연결리스트의 요소 삭제 시
                        => 마지막 노드와, 마지막 이전 노드 탐색
                        => 마지막 이전 노드의 next를 None으로 지정
                        => size 감소 (head 변화 x)
                        => 기존 마지막 노드 반환
                    else 만약 텅 빈 연결리스트라면?
                        => 비어있다고 말해주기 

    `search(key)`   if `None`을 찾아달라고 한다면? 
                        => 아무것도 안하기
                    elif 텅 빈 리스트라면? 
                        => 아무것도 안하기
                    else key값을 정상 입력 시
                        => head부터 key와 일치할 때까지
                        => 일치하지 않으면 head의 next와 비교
                        => 찾으면 해당 노드 반환
                        => 현재 노드의 next가 none이면 실패메시지 반환

    `searchTwoNode(key)->(Node, Node)`
                    => 연속한 두 노드를 동시에 찾기 (running tech)

    `delete(key)` 
        # head 노드를 지우는 경우
            1) 관리 객체의 head를 기존 head의 next로 지정
            2) size 감소
            3) 기존 head 반환
        #head가 아닌 노드를 지우는 경우
            1) 해당 요소 A와 그 이전 요소 B를 동시에 탐색 후 반환
            2) 이전 요소 B의 next를 None으로 지정
            3) size 감소
            4) 해당 요소 A 반환

        (완료) 연속된 두 키를 동시에 찾는 search 함수 구현 => 이를 삭제 연산에 활용

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.next = None
        
    def __str__(self) -> str:
        return str(self.key)
```

```python
class singleLinkedList:
    # 특수 메서드
    def __init__(self):
        self.head = None
        self.size = 0

    def __iter__(self):
        v = self.head
        while v != None:
            yield v
            v = v.next
            
    def __str__(self) -> str:
        if self.head:
            return "->".join(str(v.key) for v in self)
        else :
            return "empty"
    
    # 인스턴스 변수인 size가 올바르게 카운트 되는지 확인하기 위한, 용도로 len 사용
    def __len__(self):
        size = 0
        for v in self:
            size += 1
        return size

    # 일반 메서드
    # pushFront [v] + [v]->[]->[]->0
    def pushFront(self, key):
        #None을 키로 입력하려 한다면 아무것도 안하기
        if key == None: 
            print("값을 입력하세요 - singleLinkedList.pushFron")
        #None이 아닌 정상 키를 입력했을 때,
        else:
            newNode = Node(key)
            newNode.next = self.head #이미 요소가 있거나 없거나 잘 작동
            # 위에서 self의 요소가 Node 인스턴스를 참조하기 때문에, 
            # 현재 클래스에서 선언한 맴버변수가 Node 클래스의 맴버변수를 참조할 수 있음
            self.head = newNode 
            self.size += 1
        
    # pushBack      []->[]->[v]->0 + [v]
    def pushBack(self, key):
        #None을 키로 입력하려 한다면 아무것도 안하기
        if key == None: 
            print("값을 입력하세요 - singleLinkedList.pushBack")
            
        #None이 아닌 정상 키를 입력했을 때,
        else:
            newNode = Node(key)
            
            #빈노드면? head만 갱신
            if self.size == 0:
                self.head = newNode
                
            #비어있지 않다면? 마지막 노드 찾아서 그녀석의 next를 new로 지정
            else:
                v = self.head
                while v.next != None:
                    v = v.next
                v.next = newNode
            self.size += 1

    #key를 통해 원한는 값을 찾기 
    def search(self, key)->(Node):
        # None을 찾아달라고 하는 경우
        if key == None:
            print("값을 입력하세요 - singleLinkedList.search")
            return None
        else:
            nowNode = self.head
            # 리스트가 비어있는 경우, 혹은 모두 순회하고 현재 노드에 None이 담겼을 때 반복문 탈출
            while nowNode != None:
                # 원하는 값을 찾으면 곧바로 해당 노드 반환
                if nowNode.key == key: 
                    return nowNode
                else:
                    nowNode = nowNode.next
            print("정말 아쉽지만, 일치하는 결과가 없습니다 - singleLinkedList.search")
            return None

    #key를 통해 원한는 값을 찾기
    def searchTwoNode(self, key):
    # return -> (Node, Node)
        # None을 찾아달라고 하는 경우
        if key == None or self.size == 0: 
            print("리스트가 비어있거나, 값을 입력하지 않았습니다 - singleLinkedList.searchTwoNode")
            return None, None
        else:
            preNode, nowNode = None, self.head
            while nowNode != None:
                # 원하는 값을 찾으면 곧바로 해당 노드 반환
                if nowNode.key == key: 
                    return [preNode, nowNode]
                else:
                    preNode = nowNode
                    nowNode = nowNode.next
            print("정말 아쉽지만, 일치하는 결과가 없습니다 - singleLinkedList.searchTwoNode")
            return None, None

    def delete(self, key)->(Node):
        #none을 찾거나, 텅 빈 리스트에서 검색 하려고 시도 시
        if key == None or self.size == 0: 
            print("리스트가 비어있거나, 값을 입력하지 않았습니다 - singleLinkedList.delete")
            return None
        else:
            # Search를 호출하여, 이전 노드와 현재 노드 반환
            preNode, nowNode = self.searchTwoNode(key)
            # 요소가 둘 이상일 때 첫번째 노드를 지우는 경우, pre가 None이니까 단순히 head값만 갱신
            if preNode == None and nowNode != None:
                self.head = nowNode.next
                return nowNode
            # 탐색에 실패한 경우
            elif preNode == None and nowNode == None:
                print("해당 값을 찾을 수 없습니다 - singleLinkedList.delete")
                return None
            else:
                preNode.next = nowNode.next
                self.size -= 1
                return nowNode
        
    def popFront(self)->(Node):
        #비어있다면? 무리무리 
        if self.size == 0:
            print("더 이상 삭제할 노드가 존재하지 않습니다. - singleLinkedList.popFront")
            return None
        #비어있지 않다면? 무조건 헤드노드 갱신
        else:
            v = self.head 
            self.head = self.head.next
            self.size -= 1
            return v
        
    def popBack(self)->(Node):
        #비어있다면? 이무것도 안해
        if self.size == 0:
            print("더 이상 삭제할 노드가 존재하지 않습니다. - singleLinkedList.popBack")
            return None
        #노드가 딱 하나만 들어있다면? 삭제 후 헤드노드 갱신
        elif self.size == 1:
            v = self.head
            self.head = None
            self.size -= 1
            return v
        #노드가 둘 이상 들어있다면? 마지막 값과 이전 값 찾아서 리턴 삭제 후 갱신 불필요
        else:
            preTail, tail = self.head, self.head.next
            while tail.next != None:
                tail = tail.next
                preTail = preTail.next
            preTail.next = None
            self.size -= 1
            return tail


    def printInfo(self):
        print("    =>    info")
        print("          - List structure  | ", self)
        print("          - head:",self.head, " | size:",self.size, " | count(node):", len(self))
        print()
```
