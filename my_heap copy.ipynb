<h1>
Banary Tree
</h1>

각 요소들이 일대다 관계를 갖는 경우, 그래프라고 부르며, 특히 부모와 자식의 관계만으로 구성되어있고, 순환하지 않는 경우를 트리라고 한다. 
자식을 최대 둘까지 갖는 경우를 이진트리라고 한다.
이진트리를 코드상으로 표현할 수 있는 방법 중 세 가지를 소개한다. 

방법 1. 리스트 상에 레벨별로 나열하기 
    모든 노드가 있다고 생각하고, 각 레벨별로 모든 요소를 차례대로 표현하는 방식이다.
    빈 노드의 경우 None으로 두고, 가장 왼쪽 노드부터 오른쪽으로 적는다.
    빈 노드가 많으면 메모리 낭비가 심하다.

방법 2. 리스트에 재귀적으로 표현하기
    [자신, [왼쪽 서브트리], [오른쪽 서브트리]] 와 같은 방식을 재귀적으로 표현하는 방식이다. 

방법 3. 노드 클래스를 정의해서 이를 연결리스트와 비슷한 방식으로 연결하는 방식이다.
노드 클래스는 (key, value), 부모링크, 왼쪽자식링크, 오른쪽자식링크 이렇게 총 다섯개의 요소를 적어도 갖는다. 

<h1>
Heap Class
</h1>
힙은 이진트리 중에서도 다음의 두 가지 성질을 모두 만족하는 트리를 말한다. 

모양 성질 : 리프 노드를 제외한 모든 노드가 완전히 차있다. 마지막 레벨에선 왼쪽부터 값이 차있다.
힙 성질 : 루트를 제외한 모든 노드의 값은 자신의 부모노드보다 크면 안된다.

앞서 소개한 이진트리의 표현 방식 중에 [방법1]을 선택해서 표현한다.


- 메서드 명세
    특수 메서드
    __init__
        평범한 값을이 담긴 리스트  => 모양성질 만족 : 리스트에 담긴 값을 해석하면 모양성질을 만족한다.
        힙정렬 시키는 연산       => 힙성질 만족 : 힙성질이 만족하도록 정렬하는 과정을 거친다.

    heapifyDown             => 한 노드 제외하고, 해당 노드의 자손이 힙성질을 만족할 때, 해당 노드를 기준으로 내려가면서 힙성질을 만족하도록 재배치한다.

    insert                  => 리스트에 append하여 트리에 값을 삽입하면, 마지막 레벨의 가장 왼쪽부터 값이 쌓이며, 이를 힙성질이 만족하도록 root노드까지 올라가며 정렬해야한다.
                            => heapifyUp을 호출하여 삽입 후에 힙성질을 만족하도록 만든다.
    heapifyUp               

    findMax                 => 가장 앞에 저장된 값이 자연스럽게 최댓값이다.
    deleteMax               => 이를 삭제한 뒤엔, 다시 정렬이 필요한데 가장 작은 값을 원래 자리에 앉혀두고, 내려가며 정렬하는 방법을 쓴다.
                            => 만약 두번째로 큰 값을 올려둔다면 할일이 정말 많아진다. 
    heapSort                => 힙을 이용하여, 자료를 내림차순으로 정렬한다. 최댓값을 계속 맨 뒤로 보낸 뒤에, 해당 값이 삭제된거라고 취급하면, 값이 정렬된다
                            => 1. 임의의 리스트를 우선 힙으로 만들기 : makeHeap
                            => 2. deleteMax연산을 리스트의 크기만큼 반복
